```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)


```

```
def initialize_parameters(layer_dims):
    L = len(layer_dims)
    parameters = {}
    for l in range(1,L):
        parameters["W" + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1])*0.01
        parameters["b" + str(l)] = np.zeros((layer_dims[l],1))
    return parameters

def initialize_momentum(parameters):
    L = len(parameters)//2
    v = {}
    for l in range(L):
        v["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        v["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
    
    return v

def initialize_rms(parameters):
    L = len(parameters)//2
    s = {}
    for l in range(L):
        s["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        s["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
    
    return s

def initialize_adam(parameters):
    L = len(parameters)//2
    s = {}
    v = {}
    for l in range(L):
        v["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        v["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
        s["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        s["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
    
    return v, s

def linear_forward(A, W, b):
    Z = np.dot(W, A) + b
    cache = (A, W, b)
    return Z, cache
def softmax(Z):
    T = np.exp(Z)
    cache = Z
    T_sum = np.sum(T, axis = 0, keepdims = True)
    A = np.divide(T,T_sum)
    return A, cache
def relu(Z):
    A = np.maximum(Z,0)
    cache = Z
    return A, cache

def linear_activation_forward(A_prev, W, b, activation):
    
    if(activation == "softmax"):
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = softmax(Z)
    elif(activation == "relu"):
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = relu(Z)
    elif(activation == "tanh"):
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = np.tanh(Z), Z
    
    cache = (linear_cache, activation_cache)
    return A, cache

def forward_propagation(X, parameters, activation):
    
    caches = []
    L = len(parameters)//2
    A = X
    
    for l in range(L-1):
        A, cache = linear_activation_forward(A, parameters["W" + str(l+1)], parameters["b" + str(l+1)], activation)
        caches.append(cache)
    AL, cache = linear_activation_forward(A, parameters["W" + str(L)], parameters["b" + str(L)], "softmax")
    caches.append(cache)
    return AL, caches

def compute_cost(AL, Y):
    
    m = Y.shape[1]
    cost = -np.sum(np.multiply(Y, np.log(AL)))
    
    return cost

def linear_backward(dZ, cache):
    
    A_prev, W, b = cache
    m = A_prev.shape[1]
    dW = (1.0/m)*np.dot(dZ, A_prev.T)
    db = (1.0/m)*np.sum(dZ, axis = 1, keepdims = True)
    dA_prev = np.dot(W.T, dZ)
    return dA_prev, dW, db


def relu_backward(dA, cache):
    Z = cache
    dZ = np.array(dA, copy = True)
    dZ[Z<=0] = 0
    return dZ

def tanh_backward(dA, cache):
    Z = cache
    s = np.tanh(Z)
    dZ = dA*(1-np.power(s,2))
    return dZ

def linear_activation_backward(dA, cache, activation):
    
    (linear_cache, activation_cache) = cache
    if(activation == "softmax"):
        dZ = dA
        dA_prev, dW, db = linear_backward(dZ, linear_cache)
    elif(activation == "relu"):
        dZ = relu_backward(dA, activation_cache)
        dA_prev, dW, db = linear_backward(dZ, linear_cache)
    elif(activation == "tanh"):
        dZ = tanh_backward(dA, activation_cache)
        dA_prev, dW, db = linear_backward(dZ, linear_cache)
    return dA_prev, dW, db

def backward_propagation(AL, Y, caches, activation):
    
    grads = {}
    L = len(caches)
    
    dZL = (AL - Y)
    
    cache = caches[L-1]
    grads["dA" + str(L-1)], grads["dW" + str(L)], grads["db" + str(L)] = linear_activation_backward(dZL, cache, "softmax")
    
    for l in reversed(range(L-1)):
        cache = caches[l]
        grads["dA" + str(l)], grads["dW" + str(l+1)], grads["db" + str(l+1)] = linear_activation_backward(grads["dA" + str(l+1)], cache, activation)
    
    return grads

def update_parameters(parameters, grads, learning_rate):
    
    L = len(parameters)//2
    for l in range(L):
        parameters["W" + str(l+1)] -= learning_rate*grads["dW" + str(l+1)]
        parameters["b" + str(l+1)] -= learning_rate*grads["db" + str(l+1)]
        
    return parameters


def update_parameters_with_momentum(parameters, grads, v, learning_rate, beta = 0.9):
    
    L = len(parameters)//2
    for l in range(L):
        v["dW" + str(l+1)] = beta*v["dW" + str(l+1)] + (1-beta)*grads["dW" + str(l+1)]
        v["db" + str(l+1)] = beta*v["db" + str(l+1)] + (1-beta)*grads["db" + str(l+1)]
        
        parameters["W" + str(l+1)] -= learning_rate*v["dW" + str(l+1)]
        parameters["b" + str(l+1)] -= learning_rate*v["db" + str(l+1)]
    
    return parameters

def update_parameters_with_nesterov(parameters, grads, v, X, Y, learning_rate, activation, beta = 0.9):
    
    L = len(parameters)//2
    parameters_lookahead = {}
    for l in range(L):
        parameters_lookahead["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate*v["dW" + str(l+1)]
        parameters_lookahead["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate*v["db" + str(l+1)]
    AL_lookahead, caches = forward_propagation(X, parameters, activation)
    grads_lookahead = backward_propagation(AL_lookahead, Y, caches, activation)
    for l in range(L):
        v["dW" + str(l+1)] = beta*v["dW" + str(l+1)] + (1-beta)*grads_lookahead["dW" + str(l+1)]
        v["db" + str(l+1)] = beta*v["db" + str(l+1)] + (1-beta)*grads_lookahead["db" + str(l+1)]
        
        parameters["W" + str(l+1)] -= learning_rate*v["dW" + str(l+1)]
        parameters["b" + str(l+1)] -= learning_rate*v["db" + str(l+1)]
    return parameters

def update_parameters_with_adagrad(parameters, grads, s, learning_rate):
    L = len(parameters)//2
    for l in range(L):
        s["dW" + str(l+1)] = s["dW" + str(l+1)] + np.power(grads["dW" + str(l+1)],2)
        s["db" + str(l+1)] = s["db" + str(l+1)] + np.power(grads["db" + str(l+1)],2)
        parameters["W" + str(l+1)] -= learning_rate*np.divide(grads["dW" + str(l+1)], 1e-8 + np.sqrt(s["dW" + str(l+1)]))
        parameters["b" + str(l+1)] -= learning_rate*np.divide(grads["db" + str(l+1)], 1e-8 + np.sqrt(s["db" + str(l+1)]))
    
    return parameters

def update_parameters_with_rms(parameters, grads, s, learning_rate, t, beta = 0.999):
    s_corrected = {}
    L = len(parameters)//2
    for l in range(L):
        s["dW" + str(l+1)] = beta*s["dW" + str(l+1)] + (1-beta)*np.power(grads["dW" + str(l+1)],2)
        s["db" + str(l+1)] = beta*s["db" + str(l+1)] + (1-beta)*np.power(grads["db" + str(l+1)],2)
        parameters["W" + str(l+1)] -= learning_rate*np.divide(grads["dW" + str(l+1)], 1e-8 + np.sqrt(s["dW" + str(l+1)]))
        parameters["b" + str(l+1)] -= learning_rate*np.divide(grads["db" + str(l+1)], 1e-8 + np.sqrt(s["db" + str(l+1)]))
    
    return parameters


def update_parameters_with_adam(parameters, grads, v, s, learning_rate, t, beta1 = 0.9, beta2 = 0.999):
    s_corrected = {}
    v_corrected = {}
    L = len(parameters)//2
    for l in range(L):
        v["dW" + str(l+1)] = beta1*v["dW" + str(l+1)] + (1-beta1)*grads["dW" + str(l+1)]
        v["db" + str(l+1)] = beta1*v["db" + str(l+1)] + (1-beta1)*grads["db" + str(l+1)]
        v_corrected["dW" + str(l+1)] = v["dW" + str(l+1)]/(1-beta1**t)
        v_corrected["db" + str(l+1)] = v["db" + str(l+1)]/(1-beta1**t)
        s["dW" + str(l+1)] = beta2*s["dW" + str(l+1)] + (1-beta2)*np.power(grads["dW" + str(l+1)],2)
        s["db" + str(l+1)] = beta2*s["db" + str(l+1)] + (1-beta2)*np.power(grads["db" + str(l+1)],2)
        s_corrected["dW" + str(l+1)] = s["dW" + str(l+1)]/(1-beta2**t)
        s_corrected["db" + str(l+1)] = s["db" + str(l+1)]/(1-beta2**t)
        parameters["W" + str(l+1)] -= learning_rate*np.divide(v_corrected["dW" + str(l+1)], 1e-8 + np.sqrt(s_corrected["dW" + str(l+1)]))
        parameters["b" + str(l+1)] -= learning_rate*np.divide(v_corrected["db" + str(l+1)], 1e-8 + np.sqrt(s_corrected["db" + str(l+1)]))
    
    return parameters

def random_mini_batches(X, Y, mini_batch_size = 64):
    
    mini_batches = []
    m = X.shape[1]
    permutations = list(np.random.permutation(m))
    shuffled_X = X[:, permutations]
    shuffled_Y = Y[:, permutations]
    
    num_complete_minibatches = int(m/mini_batch_size)
    for i in range(num_complete_minibatches):
        mini_batch_X = shuffled_X[:, i*(mini_batch_size): (i+1)*(mini_batch_size)]
        mini_batch_Y = shuffled_Y[:, i*(mini_batch_size): (i+1)*(mini_batch_size)]
        mini_batch = (mini_batch_X, mini_batch_Y)
        mini_batches.append(mini_batch)
        
    if i%mini_batch_size != 0:
        mini_batch_X = shuffled_X[:, num_complete_minibatches*(mini_batch_size): m]
        mini_batch_Y = shuffled_Y[:, num_complete_minibatches*(mini_batch_size): m]
        mini_batch = (mini_batch_X, mini_batch_Y)
        mini_batches.append(mini_batch)
    
    return mini_batches

def NN_model_optimizer(X, Y, layer_dims, max_iter = 100, mini_batch_size = 64, solver = "gd",  learning_rate = 0.001, beta = 0.9, beta1 = 0.9, beta2 = 0.999, activation = "relu"):
    parameters = initialize_parameters(layer_dims)
    
    if solver == "momentum" or solver == "nerestov":
        v = initialize_momentum(parameters)
    elif solver == "rms" or solver == "adagrad":
        s = initialize_rms(parameters)
    elif solver == "adam":
        v, s = initialize_adam(parameters)
    
    costs = []
    m = X.shape[1]
    t = 1
    for i in range(max_iter):
        mini_batches = random_mini_batches(X, Y, mini_batch_size)
        cost = 0
        for mini_batch in mini_batches:
            
            (mini_batch_X, mini_batch_Y) = mini_batch
            
            AL, caches = forward_propagation(mini_batch_X, parameters, activation)
            cost += compute_cost(AL, mini_batch_Y)
            grads = backward_propagation(AL, mini_batch_Y, caches, activation)
            
            if solver == "gd":
                parameters = update_parameters(parameters, grads, learning_rate)
            elif solver == "momentum":
                parameters = update_parameters_with_momentum(parameters, grads, v, learning_rate, beta)
            elif solver == "nerestov":
                parameters = update_parameters_with_nesterov(parameters, grads, v, mini_batch_X, mini_batch_Y, learning_rate, activation, beta)
            elif solver == "rms":
                parameters = update_parameters_with_rms(parameters, grads, s, learning_rate, beta)
            elif solver == "adagrad":
                parameters = update_parameters_with_adagrad(parameters, grads, s, learning_rate)
            elif solver == "adam":
                parameters = update_parameters_with_adam(parameters, grads, v, s, learning_rate, t,beta1, beta2)
                t = t + 1
        if i%50 == 0:
            print("Cost at iteration {} = {}".format(i, cost / m))
        
        costs.append(cost / m)
    
    plt.plot(range(len(costs)), costs, label = solver)
    return parameters

def predict(X, parameters, activation):
    AL, _ = forward_propagation(X, parameters, activation)
    Y_pred = np.zeros(AL.shape)
    for i,a in zip(range(AL.shape[1]), np.argmax(AL, axis = 0)):
        Y_pred[int(a), i] = 1
    return Y_pred
```

```
f=open("connect-4.data",'r')
x1=f.readlines()
#print(x1)
f.close()
n=len(x1)
x2=[]
for i in range(n):
    x2.append(x1[i].rstrip().split(','))
data = np.array(x2)
data_flat = data.ravel()
my_dict = {'b':0, 'x':1, 'o':2, 'win':0, 'loss':1, 'draw':2}

data_encoded = np.array([my_dict[i] for i in data_flat], np.float32).reshape(data.shape)
#clf = OrdinalEncoder().fit(data)
#data_encoded = clf.transform(data)
Y_direct = data_encoded[:,data_encoded.shape[1]-1].reshape((data.shape[0],1))
X = data_encoded[:,0:data_encoded.shape[1]-1].reshape((data_encoded.shape[0], data_encoded.shape[1]-1))

Y = np.zeros((X.shape[0], 3))
for i,y in zip(range(Y_direct.shape[0]),Y_direct):
    Y[i, int(y)] = 1
#Y[:, list(Y_direct.astype(int))] = 1.0

indices = np.arange(X.shape[0])
np.random.shuffle(indices)
num_training_indices = int(0.75 * X.shape[0])
train_indices = indices[:num_training_indices]
test_indices = indices[num_training_indices:]
X_train, X_test = X[train_indices].T, X[test_indices].T
Y_train, Y_test = Y[train_indices].T, Y[test_indices].T

np.random.seed(1)
```

```
layer_dims = [X_train.shape[0], 100, 10, Y_train.shape[0]]
solvers = ["gd", "momentum", "nerestov", "adagrad","rms", "adam"]
solvers_name = ["Gradient descent", "Momentum","Nerestov", "Adagrad", "RMSprop", "Adam"]
plt.figure()
for i in range(len(solvers)):
    parameters = NN_model_optimizer(X_train, Y_train, layer_dims, max_iter = 100, solver = solvers[i], mini_batch_size = 64, learning_rate = 1e-2)
    train_acc = np.mean(np.argmax(predict(X_train, parameters, "relu"), axis = 0) == np.argmax(Y_train, axis = 0))
    test_acc = np.mean(np.argmax(predict(X_test, parameters, "relu"), axis = 0) == np.argmax(Y_test, axis = 0))
    print("Train Accuracy using {} = {}%".format(solvers_name[i], train_acc*100) )
    print("Test Accuracy using {} = {}%".format(solvers_name[i], test_acc*100) )

plt.grid()
plt.legend()
plt.show()
```
